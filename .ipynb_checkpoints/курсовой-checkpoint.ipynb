```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import re

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
data_load = pd.read_csv('bbc-news-data.csv', sep='\t')
data_load
```

```python
data_load['category'].unique()
```

```python
def transform(data_load, test_size=0.2):
    data = data_load.copy()
    data.replace({'category':{'business':1, 'entertainment':2, 'politics':3, 'sport':4, 'tech':5}}, inplace=True)
    target = data['category']
    data = data[['title', 'content']]
#     data = data['title'].str.lower()
    if test_size == 0:
        return data, target
    x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=test_size)
    return x_train, x_test, y_train, y_test
    
# transform(data_load)
```

```python
def clean_text(text):
    if not isinstance(text, str):
        text = str(text)
    
    text = text.lower()
    text = text.strip('\n').strip('\r').strip('\t')
    text = re.sub("-\s\r\n\|-\s\r\n|\r\n", '', str(text))

    text = re.sub("[0-9]|[-—.,:;_%©«»?*!@#№$^•·&()]|[+=]|[[]|[]]|[/]|", '', text)
    text = re.sub(r"\r\n\t|\n|\\s|\r\t|\\n", ' ', text)
    text = re.sub(r'[\xad]|[\s+]', ' ', text.strip())
    return text

```

```python
x_train, x_test, y_train, y_test = transform(data_load)
x_train
```

```python
class That(BaseEstimator, TransformerMixin):
    def __init__(self, this = True):
        self.this = this
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        return X[self.this].apply(lambda x: clean_text(x), 1)
#         return X[self.this]


pip = Pipeline([('sdfs', That('content')), 
                ('dasd', TfidfVectorizer(max_features=5000, stop_words='english'))])

# resdd = ColumnTransformer(transformers=[('df', pip(), ['title', 'content'])])
    
# pi = Pipeline([('sdf', resdd()), ('dfd', LogisticRegression())])
# pi.fit(x_train, y_train)

a = pip.fit_transform(data_load)
a.toarray().shape
```
